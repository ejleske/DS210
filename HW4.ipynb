```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
# defining the data
# Define the function f
def f(x):
    return 0 if x < 0.5 else 1
# Generate sample data
np.random.RandomState(0)
X = np.linspace(0, 1, 1000)
y = 1 / (1 + np.exp(-10 * (X - 0.5)))
# Add some noise to data 
y_noise = y + np.random.normal(0, 0.1, size=X.shape)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1, 1), y_noise, test_size=0.2, random_state=42)
```

```python
# Train decision trees with different loss functions
tree_mse = DecisionTreeRegressor(max_leaf_nodes=6, criterion='squared_error')
tree_mae = DecisionTreeRegressor(max_leaf_nodes=6, criterion='absolute_error')
tree_mse.fit(X_train, y_train)
tree_mae.fit(X_train, y_train)


#predicting the values for mean square error and mean absolute error
y_1 = tree_mse.predict(X.reshape(-1, 1))
y_2 = tree_mae.predict(X.reshape(-1, 1))
# Plot the original function and the predicted functions
plt.figure(figsize=(10, 6))
plt.scatter(X, y_noise, s=20, c="yellow", label="data")
#plt.plot(X, y, color="green", label="True function", linewidth=2)
plt.plot(X, y_1, color="blue", label="Predicted (MSE)", linewidth=2, linestyle='--')
plt.plot(X, y_2, color="red", label="Predicted (MAE)", linewidth=2, linestyle='--')
plt.xlabel("Data")
plt.ylabel("Target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
